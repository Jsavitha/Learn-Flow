#  QUIZ GAME

```python
import random
```

```python
class ProgrammingQuiz:
    def __init__(self):
        self.programming_topics = {
            'Python Programming': [
                {
                    'question': 'What is the purpose of the "len()" function in Python?',
                    'options': ['Calculate length of a list', 'Convert to lowercase', 'Generate a random number', 'Open a file'],
                    'correct_answer': 'Calculate length of a list'
                },
                {
                    'question': 'What is the difference between "==" and "is" in Python?',
                    'options': ['Used for comparison', 'Used for identity testing', 'Used for assignment', 'Used for bitwise AND'],
                    'correct_answer': 'Used for comparison'
                },
                {
                    'question': 'How do you comment multiple lines in Python?',
                    'options': ['// Comment //', '/* Comment */', '# Comment', '-- Comment'],
                    'correct_answer': '# Comment'
                }
            ],
            'C Programming': [
                {
                    'question': 'What is the syntax for declaring a pointer in C?',
                    'options': ['int *ptr;', 'int ptr;', 'pointer ptr;', 'ptr *int;'],
                    'correct_answer': 'int *ptr;'
                },
                {
                    'question': 'Which function is used to allocate memory in C?',
                    'options': ['alloc()', 'malloc()', 'new()', 'allocate()'],
                    'correct_answer': 'malloc()'
                },
                {
                    'question': 'What is the purpose of the "sizeof" operator in C?',
                    'options': ['Calculate size of data type', 'Find the length of a string', 'Allocate memory', 'Determine the size of a file'],
                    'correct_answer': 'Calculate size of data type'
                }
            ],
            'C++ Programming': [
                {
                    'question': 'What is an object in C++?',
                    'options': ['A variable', 'A function', 'An instance of a class', 'A loop'],
                    'correct_answer': 'An instance of a class'
                },
                {
                    'question': 'Which keyword is used for inheritance in C++?',
                    'options': ['inherit', 'extends', 'inheritance', 'class'],
                    'correct_answer': 'inheritance'
                },
                {
                    'question': 'What is the purpose of the "virtual" keyword in C++?',
                    'options': ['Define a constant', 'Enable dynamic binding', 'Declare a function', 'Allocate memory'],
                    'correct_answer': 'Enable dynamic binding'
                }
            ],
            # Add more programming topics as needed
        }
        self.quiz_score = 0

    # Displaying a Question
    def display_question(self, programming_question):
        print(programming_question['question'])
        for i, option in enumerate(programming_question['options'], start=1):
            print(f"{i}. {option}")
        user_answer = input("Enter the number of your answer: ")
        return user_answer

    # Checking the Answer
    def check_answer(self, programming_question, user_answer):
        correct_answer = programming_question['options'].index(programming_question['correct_answer']) + 1
        if user_answer == str(correct_answer):
            print("Correct! Well done.")
            self.quiz_score += 1
        else:
            print(f"Sorry, the correct answer is {programming_question['correct_answer']}.")

    # Starting the Quiz
    def start_quiz(self):
        print("Welcome to the Programming Quiz Game!")

        # Choose a programming language topic
        print("\nChoose a programming language topic:")
        for i, topic in enumerate(self.programming_topics.keys(), start=1):
            print(f"{i}. {topic}")

        topic_choice = int(input("Enter the number of your chosen topic: "))
        selected_topic = list(self.programming_topics.keys())[topic_choice - 1]
        questions_to_ask = random.sample(self.programming_topics[selected_topic], min(3, len(self.programming_topics[selected_topic])))

        for question in questions_to_ask:
            user_answer = self.display_question(question)
            self.check_answer(question, user_answer)

        print(f"\nQuiz completed! Your score is: {self.quiz_score}/{len(questions_to_ask)}")

# Main Execution
if __name__ == "__main__":
    programming_quiz = ProgrammingQuiz()
    programming_quiz.start_quiz()
```
